# t-SNE

```python
# imports and configs
import os
import pandas as pd
from sklearn.manifold import TSNE
import seaborn as sns

PROJECT_ROOT = f"{os.path.expanduser('~')}/project/AIAA5030_Data_Mining_Group13_Visualization"
DATA_ROOT = f"{PROJECT_ROOT}/data"
FIGURE_ROOT = f"{PROJECT_ROOT}/figure"
```

## Titanic Dataset
Visualizing Titanic dataset using t-SNE for dimensionality reduction and matplotlib for plotting.

```python
# load dataset
Titanic_dataset = pd.read_csv(f'{DATA_ROOT}/Titanic-Dataset-Processed.csv')
Titanic_dataset
```

```python
t_SNE = TSNE(n_components=2, random_state=42, perplexity=5)
# Pclass, Sex, Age, SibSp, Parch, Fare, and Embarked
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
Titanic_t_SNE = t_SNE.fit_transform(Titanic_dataset[features])
df = pd.DataFrame(data=Titanic_t_SNE, columns=['t-SNE Component 1', 't-SNE Component 2'])
df['Survived'] = Titanic_dataset['Survived']

t_SNE_img = sns.scatterplot(data=df, x='t-SNE Component 1', y='t-SNE Component 2', hue='Survived', palette='Set1', s=10)
t_SNE_img.figure.savefig(f'{FIGURE_ROOT}/t-SNE_Titanic_Dataset_Survived_perplexity5.png', dpi=600)
t_SNE_img
```
